```python
import os
import pandas as pd
import glob
import numpy as np
os.getcwd()
```

```python
path = r'./gridsearchresult' # use your path
all_files = glob.glob(path + "/*.csv")
bestHyperDf = pd.DataFrame()
for filename in all_files:
    newDf = pd.read_csv(filename, index_col=None, header=0)
    newDf =(newDf[newDf['loss']==newDf['loss'].min()])
    newDf['files'] = str(filename.rsplit('/', 1)[1])
    bestHyperDf = bestHyperDf.append(newDf)
#bestHyperDf.drop(['Unnamed: 0'],axis=1,inplace=True)
```
